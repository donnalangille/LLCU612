## Mini-Assignment 1







This assignment is to generate a list of current headlines from CBC and FoxNews. 

First, I will request the source code from FoxNews.

```python
import urllib.request
from bs4 import BeautifulSoup

foxSource = urllib.request.urlopen('https://www.foxnews.com').read()
```

Next, I will generate headlines from FoxNews using BeautifulSoup. 

```python
foxSoup = BeautifulSoup(foxSource, 'lxml')
foxArticles = foxSoup.find_all('article')
for foxArticle in foxArticles:
    foxHeader = foxArticle.find("h2")
    if foxHeader:
        foxAnchor = foxHeader.find("a")
        print(foxAnchor.string)
```

Now I'm going to try generating the headlines from CBC. First, I will request the source code.

```python
import urllib.request
response = urllib.request.urlopen('https://www.cbc.ca')
html = response.read()
```

Next, I will generate the headlines using Beautiful Soup. 

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'lxml')
```

```python
for headline in soup.find_all("h3", class_="headline"):
    print(headline.text)
```
